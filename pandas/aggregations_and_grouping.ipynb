```python
country_respondents = df["Country"].value_counts()
country_uses_python = country_group["LanguageWorkedWith"].apply(
    lambda series: series.str.contains("Python").sum()
)
```

```python
python_df = pd.concat(
    [country_respondents, country_uses_python],
    axis="columns",
)
python_df
```

```python
python_df.rename(
    columns={"Country": "NumRespondents", "LanguageWorkedWith": "NumKnowPython"},
    inplace=True
)
```

```python
python_df["PctKnowsPython"] = round(
    (python_df["NumKnowPython"] / python_df["NumRespondents"]) * 100, 2
)
python_df
```

```python
python_df.sort_values(by="PctKnowsPython", ascending=False)
```
